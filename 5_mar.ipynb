```python
#3
class node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to detect a cycle in the linked list
def detect_cycle(head):
    slow = head  # Slow pointer (tortoise)
    fast = head  # Fast pointer (hare)
    
    while fast and fast.getnext():  # Move until the fast pointer or its next is None
        slow = slow.getnext()       # Move slow pointer by 1 step
        fast = fast.getnext().getnext()  # Move fast pointer by 2 steps
        
        if slow == fast:  # If slow and fast meet, there's a cycle
            return True
    
    return False  # If fast reaches None, there's no cycle

# Create the linked list with a cycle for testing
head = node(1)
a = node(2)
b = node(3)
c = node(4)
d = node(5)

head.setnext(a)
a.setnext(b)
b.setnext(c)
c.setnext(d)
d.setnext(b)  # Creating a cycle (pointing back to node `b`)

# Detect cycle
if detect_cycle(head):
    print("Cycle detected!")
else:
    print("No cycle detected.")
```

```python
#4
class node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to merge two sorted linked lists
def merge_sorted_lists(l1, l2):
    dummy = node(0)  # Dummy node to help build the result list
    tail = dummy  # Tail pointer to track the end of the result list
    
    while l1 and l2:
        if l1.getdata() <= l2.getdata():
            tail.setnext(l1)  # Attach l1 node
            l1 = l1.getnext()  # Move l1 pointer
        else:
            tail.setnext(l2)  # Attach l2 node
            l2 = l2.getnext()  # Move l2 pointer
        tail = tail.getnext()  # Move the tail to the new end
    
    # If any nodes remain in either list, attach them
    if l1:
        tail.setnext(l1)
    else:
        tail.setnext(l2)
    
    return dummy.getnext()  # Return the next node after dummy, which is the head of the merged list

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.getdata(), end=" ")
        current = current.getnext()
    print()

# Creating two sorted linked lists
# List 1: 1 -> 3 -> 5
head1 = node(1)
head1.setnext(node(3))
head1.getnext().setnext(node(5))

# List 2: 2 -> 4 -> 6
head2 = node(2)
head2.setnext(node(4))
head2.getnext().setnext(node(6))

# Merging the lists
merged_head = merge_sorted_lists(head1, head2)

# Printing the merged linked list
print("Merged Sorted Linked List:")
print_list(merged_head)
```

```python
#5
class node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to remove the nth node from the end of the linked list
def remove_nth_from_end(head, n):
    dummy = node(0)  # Create a dummy node to handle edge cases
    dummy.setnext(head)
    
    fast = dummy
    slow = dummy
    
    # Move fast pointer n+1 steps ahead so the gap between slow and fast is n nodes
    for _ in range(n + 1):
        fast = fast.getnext()
    
    # Move both slow and fast pointers together until fast reaches the end
    while fast:
        slow = slow.getnext()
        fast = fast.getnext()
    
    # Slow is now at the (n+1)th node from the end, remove the nth node
    slow.setnext(slow.getnext().getnext())
    
    return dummy.getnext()  # Return the new head

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.getdata(), end=" ")
        current = current.getnext()
    print()

# Creating a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = node(1)
head.setnext(node(2))
head.getnext().setnext(node(3))
head.getnext().getnext().setnext(node(4))
head.getnext().getnext().getnext().setnext(node(5))

# Remove the 2nd node from the end
n = 2
new_head = remove_nth_from_end(head, n)

# Print the new linked list
print("Linked List after removing the 2nd node from the end:")
print_list(new_head)
```

```python
#6
class node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to remove duplicates from a sorted linked list
def remove_duplicates(head):
    current = head
    
    while current and current.getnext():  # Traverse the list
        if current.getdata() == current.getnext().getdata():  # If duplicate found
            current.setnext(current.getnext().getnext())  # Skip the duplicate node
        else:
            current = current.getnext()  # Move to the next node if no duplicate
    
    return head

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.getdata(), end=" ")
        current = current.getnext()
    print()

# Creating a sorted linked list: 1 -> 1 -> 2 -> 3 -> 3 -> 4 -> 4 -> 4 -> 5
head = node(1)
head.setnext(node(1))
head.getnext().setnext(node(2))
head.getnext().getnext().setnext(node(3))
head.getnext().getnext().getnext().setnext(node(3))
head.getnext().getnext().getnext().getnext().setnext(node(4))
head.getnext().getnext().getnext().getnext().getnext().setnext(node(4))
head.getnext().getnext().getnext().getnext().getnext().getnext().setnext(node(4))
head.getnext().getnext().getnext().getnext().getnext().getnext().getnext().setnext(node(5))

# Remove duplicates
new_head = remove_duplicates(head)

# Print the new linked list
print("Linked List after removing duplicates:")
print_list(new_head)
```

```python
#8
class node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to find the length of the linked list
def get_length(head):
    length = 0
    current = head
    while current:
        length += 1
        current = current.getnext()
    return length

# Function to rotate the linked list by k positions to the right
def rotate_right(head, k):
    if not head or not head.getnext() or k == 0:
        return head
    
    # Step 1: Find the length of the linked list
    length = get_length(head)
    
    # Step 2: Normalize k (if k >= length)
    k = k % length
    if k == 0:
        return head  # No rotation needed
    
    # Step 3: Find the new tail (the (length - k)th node) and new head
    current = head
    for _ in range(length - k - 1):
        current = current.getnext()
    
    # Step 4: Set the new head and tail
    new_head = current.getnext()
    current.setnext(None)  # Break the link at the new tail
    
    # Step 5: Find the end of the new list and connect it to the old head
    tail = new_head
    while tail.getnext():
        tail = tail.getnext()
    tail.setnext(head)  # Connect the end of the rotated list to the old head
    
    return new_head

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.getdata(), end=" ")
        current = current.getnext()
    print()

# Creating a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = node(1)
head.setnext(node(2))
head.getnext().setnext(node(3))
head.getnext().getnext().setnext(node(4))
head.getnext().getnext().getnext().setnext(node(5))

# Rotate the list by k positions
k = 2
new_head = rotate_right(head, k)

# Print the rotated linked list
print(f"Linked List after rotating by {k} positions:")
print_list(new_head)
```

```python
#9
class node:
    def __init__(self, data=0, next=None):
        self.data = data
        self.next = next
    def setdata(self, data):
        self.data = data
    def getdata(self):
        return self.data
    def setnext(self, node):
        self.next = node
    def getnext(self):
        return self.next

# Function to add two numbers represented by linked lists
def add_two_numbers(l1, l2):
    dummy = node(0)  # Dummy node to start the result list
    current = dummy  # Pointer to build the result list
    carry = 0  # To store the carry over from each sum
    
    # Traverse both linked lists
    while l1 or l2 or carry:
        # Get the values from both lists, or 0 if one list is shorter
        val1 = l1.getdata() if l1 else 0
        val2 = l2.getdata() if l2 else 0
        
        # Calculate the sum of both digits plus any carry
        total = val1 + val2 + carry
        carry = total // 10  # Update carry for next iteration
        current.setnext(node(total % 10))  # Create a new node with the digit part
        
        # Move to the next nodes in l1, l2, and the result list
        current = current.getnext()
        if l1:
            l1 = l1.getnext()
        if l2:
            l2 = l2.getnext()
    
    return dummy.getnext()  # Return the result list starting after the dummy node

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.getdata(), end=" ")
        current = current.getnext()
    print()

# Creating two linked lists that represent the numbers 342 (reverse order 2 -> 4 -> 3)
# and 465 (reverse order 5 -> 6 -> 4)
l1 = node(2)
l1.setnext(node(4))
l1.getnext().setnext(node(3))

l2 = node(5)
l2.setnext(node(6))
l2.getnext().setnext(node(4))

# Add the two numbers
result = add_two_numbers(l1, l2)

# Print the result linked list
print("Sum of the two numbers:")
print_list(result)
```
